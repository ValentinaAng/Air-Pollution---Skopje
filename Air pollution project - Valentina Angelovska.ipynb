# Skopje Air Pollution

## Introduction

The city of Skopje is well known for its struggles with poor air quality, especially in the winter months. The main pollutants are CO, NO2, ozone, and particulate matter of different dimensions.

Monitoring stations for measuring the concentrations of each of the main pollutants have been installed in different locations around the city.

## Question 7
On how many days in each year, was the value of PM10 measured on at least one location worse than moderate? .......... (3 points)

```python
# cekor 1: od celiot df filtriram da mi gi dade podatocite samo za PM10 i stavam index kolona da mi bide date i year za posle da
# mozam da grupiram oti inaku mi dava deka datatype datetime64[ns] nemozi da se sporedi so int (>154)
PM10 = df[(df['type_of_pollutant']=='PM_10')].set_index(['date','year'])

# cekor 2: gi zemam site redovi od toj df i prvite 6 koloni kade vrednosta mi e pogolema od 154, grupiram po date i year 
# i pravam count na tie denovi kade shto se srejkava value > 154
new_df_PM10 = PM10[PM10.iloc[:,:6]> 154].reset_index().groupby(['date','year']).count()

#cekor 3: dodavam kolona sum koja ke mi sumira na toj datum na taa lokacija kolku pati mi bila vrednosta >154
new_df_PM10['sum'] = new_df_PM10.sum(axis=1)

#cekor4: zemam samo denovite kade vrednosta mi e pogolema od 0, i grupiram po year pa brojam kolku denovi bile vo taa godina
count_days_in_each_year = new_df_PM10[new_df_PM10['sum']>0].groupby('year').count()

#zemam od toj df da mi ja dade samo kolonata sum
result = count_days_in_each_year['sum']

result.reset_index()
```

## Question 8
Make a bar plot showing the average number of days per year when the measured value for PM10 was worse than moderate on at least one location.  .......... (3 points)

```python
result = result.reset_index() # resetiram index
```

```python
# definiram barplot 
ax = sns.barplot(data = result,x='year',y='sum', hue = 'sum',dodge=False, palette = sns.color_palette("Greens_d", len(result)))

#bar data values:
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.1399, p.get_height() * 1.005))

# definiranje na iminja na grafik, x i y oska 
plt.xlabel('Year')
plt.ylabel('Average number of days per year')
plt.title('Average Number of Days per Year when PM10 > 154')
plt.ylim(0,250) 

# definiranje na legendata
ax.legend(title = 'number of days',loc='center left', bbox_to_anchor=(1.0, 0.5))

#iscrtuvanje na grafik
plt.show()
```
